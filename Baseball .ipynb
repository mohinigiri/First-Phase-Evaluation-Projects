```python
maxaccu = 0
maxrc = 0

for i in range(1,250):
    X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=i)
    LR = LinearRegression()
    LR.fit(X_train,y_train)
    predictions = LR.predict(X_test)
    acc = r2_score(y_test,predictions)
    if acc > maxaccu:
        maxaccu=acc
        maxrc=i
print('max acc is',maxaccu,'on r state',maxrc)    
        
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=maxrc)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
y_predO = regressor.predict(X_test)
pred_train =  regressor.predict(X_train)
print('r2_score',r2_score(y_test,y_predO))
print('r2_score on trining data',r2_score(y_train,pred_train)*100)
print('mean_absolute_error',mean_absolute_error(y_test,y_predO))
print('mean_squared_error',mean_squared_error(y_test,y_predO))
print('root mean_squared_error',np.sqrt(mean_squared_error(y_test,y_predO)))
```

```python
from sklearn.model_selection import cross_val_score
```

```python
score = cross_val_score(regressor,x,y)
print(score)
print(score.mean())
print('difrence-',(r2_score(y_test,y_predO)-score.mean())*100)
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV
```

```python
param_grid = {
    'copy_X': [True,False],
    'fit_intercept': [True,False],
    'n_jobs': [None,0,1,10,20],
    'positive': [True,False],
     }
```
